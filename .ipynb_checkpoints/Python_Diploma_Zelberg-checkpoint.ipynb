Делаем более короткий датафрейм - только с количественными данными для получения статистических параметров

```python
hr_short = hr_satisf.filter(items = ['satisfaction_level', 'last_evaluation', 'number_project', 'average_montly_hours', 
                           'time_spend_company'])
hr_short.head()
```

Первый способ получения статистики:
Рассчитан каждый показатель отдельно методами numpy или pandas, результат статистики по столбам перевернут в сводную таблицу, затем результаты всех статистик добавлены в один датафрейм. Не устроило, что pivot_table расставляет столбцы в результирующей таблице по алфавиту, мне бы хотелось оставить порядок столбцов таким же, как в исходной таблице.

```python
means = np.mean(hr_satisf[['satisfaction_level', 'last_evaluation', 'number_project', 'average_montly_hours', 
                           'time_spend_company']]).reset_index().pivot_table(columns = 'index').rename({0: 'mean'})
medians = hr_satisf[['satisfaction_level', 'last_evaluation', 'number_project', 'average_montly_hours', 
                     'time_spend_company']].median().reset_index().pivot_table(columns = 'index').rename({0: 'median'})
modes = hr_satisf[['satisfaction_level', 'last_evaluation', 'number_project', 
                     'average_montly_hours', 'time_spend_company']].mode().rename({0: 'mode 1', 1: 'mode 2'})
mins = np.min(hr_satisf[['satisfaction_level', 'last_evaluation', 'number_project', 'average_montly_hours', 
                         'time_spend_company']]).reset_index().pivot_table(columns = 'index').rename({0: 'min'})
maxs = np.max(hr_satisf[['satisfaction_level', 'last_evaluation', 'number_project', 'average_montly_hours', 
                         'time_spend_company']]).reset_index().pivot_table(columns = 'index').rename({0: 'max'})
stds = hr_satisf[['satisfaction_level', 'last_evaluation', 'number_project', 'average_montly_hours', 
                  'time_spend_company']].std().reset_index().pivot_table(columns = 'index').rename({0: 'std'})
pd.concat([means, medians, modes, mins, maxs, stds])
```

Второй способ получения статистики:
Рассчитан каждый показатель и транспонирован в горизонтальную таблицу, затем добавлены строки в один датафрейм

```python
#Второй способ представления статистики
means_ = np.mean(hr_short).reset_index().transpose().rename({0: 'mean'})
means__ = means_.rename(columns=means_.iloc[0]).drop('index')
#means__

medians_ = hr_short.median().reset_index().transpose().rename({0: 'median'})
medians__ = medians_.rename(columns=medians_.iloc[0]).drop('index')
#medians__

mins_ = np.min(hr_short).reset_index().transpose().rename({0: 'min'})
mins__ = mins_.rename(columns=mins_.iloc[0]).drop('index')
#mins__

maxs_ = np.max(hr_short).reset_index().transpose().rename({0: 'max'})
maxs__ = maxs_.rename(columns=maxs_.iloc[0]).drop('index')
#maxs__

stds_ = hr_short.std().reset_index().transpose().rename({0: 'std'})
stds__ = stds_.rename(columns=stds_.iloc[0]).drop('index')
#stds__

modes_ = hr_short.mode().rename({0: 'mode 1', 1: 'mode 2'})
#modes_

statistics = means__.append(medians__).append(modes_).append(mins__).append(maxs__).append(stds__)
statistics
```

2. Last_evaluation - Время с момента последней оценки в годах. Среднее и медиана почти совпадают: 0,72 года, или 8,6 месяцев. Обязательная оценка проводится 1 раз в год.
3. Number_projects - Количество проектов, выполненных за время работы. В среднем это 3,8 проекта, минимум - 2, максимум - 7 проектов.
4. Average_monthly_hours - Среднее количество часов на рабочем месте в месяц. В среднем это 201 час в месяц, или 9,6 часов в день при 21 рабочих днях в месяце, минимум - 96 часов, максимум - 310 часов, т.е. это не менее 10 часов в день без выходных или 14,5 часов в день при 21 рабочих днях в месяце.

### 4 Рассчитайте сколько сотрудников работает в каждом департаменте. 5

Делаем выборку только тех сотрудников, которые работают (не уволены).
Применяем value_counts, переопределяем в дата фрейм с именем столбца - employees (количество сотрудников)

```python
hr_not_left = hr_satisf[hr_satisf.left == 0]
hr_employees_by_department =  hr_not_left.department.value_counts().reset_index().rename(columns={'index': 'department', 
                                                                                          'department': 'employees'})
hr_employees_by_department
```

Другой способ - делаем группировку в датафрейме по департаментам, считаем значения, сортируем по убыванию, 
переименовываем столбец.
Первый способ, через value_count, выглядит короче в записи, возможно, быстрее, т.к. сортирует по убыванию счетчика сразу.

```python
hr_employees_by_department_2 = hr_not_left.groupby(by = ['department']).count()[['satisfaction_level']].sort_values(
    by=['satisfaction_level'], ascending = False).rename(columns={'satisfaction_level': 'employees'})
hr_employees_by_department_2
```

```python
# Добавлен столбец с расчетом, какую долю в % занимает численность каждого департамента от общей численности работающих
hr_employees_by_department_2['share'] = (hr_employees_by_department_2[['employees']] / hr_not_left.count()[
    'satisfaction_level'] * 100).round(2)
hr_employees_by_department_2
```

```python
# Считаем среднее для двух выборок
hr_high_salary_hours.mean(), hr_low_salary_hours.mean()
```

Выдвигаем нулевую гипотезу Н0 для выборки с низкими зарплатами при их среднем кол-ве часов работы - 201 час в месяц: 
у сотрудников с высокими зарплатами среднее кол-во часов работы больше - 210 часов в месяц.

```python
# Считаем t-статистику и p-value по методу
from scipy.stats import stats
stats.ttest_1samp(a=hr_low_salary_hours,popmean=210)
```

Получили значение p-value ничтожно малое - 4.50578891e-51, т.е. мы отвергаем нулевую гипотезу: не можем признать, что 
время работы сотрудников с высокими зарплатами больше, чем у сотрудников с низкими зарплатами.

```python
# Считаем ту же самую нулевую гипотезу, но применяем к ряду часов работы с высокими зарплатами. 
stats.ttest_1samp(a=hr_high_salary_hours,popmean=190)
```

Результат применения ttest для ряда часов работы с высокими зарплатами получается тот же: 
p-value очень маленькое, нулевую гипотезу отвергаем.

```python
# Считаем ту же самую нулевую гипотезу, но значение для сравнения рассчитываем сразу как среднее минус 10 часов
# Результат такой же: p-value очень маленькое, нулевую гипотезу отвергаем
stats.ttest_1samp(a=hr_high_salary_hours,popmean=(hr_high_salary_hours.mean()-10))
```

Проверим обратное. Сформулируем нулевую гипотез по-другому: работники с низкими зарплатами проводят на работе больше времени, чем работники с высокими зарплатами.

```python
stats.ttest_1samp(a=hr_high_salary_hours,popmean=(hr_high_salary_hours.mean()+10))
```

```python
stats.ttest_1samp(a=hr_high_salary_hours,popmean=(hr_high_salary_hours.mean()+3))
```

```python
stats.ttest_1samp(a=hr_high_salary_hours,popmean=(hr_high_salary_hours.mean()+1))
```

```python
stats.ttest_1samp(a=hr_high_salary_hours,popmean=(hr_high_salary_hours.mean()-1))
```

#### Вывод по п. 7:
Нулевая гипотеза, что сотрудники с высоким окладом проводят на работе больше времени, чем сотрудники с низким окладом опровергнута.
Примечание:
Если сравниваем выборку с высокими зарплатами со средним кол-вом часов работы на 3, 5 10 больше или меньше, чем есть на самом деле, то получаем, что такую гипотезу отвергаем. Если сравниваем со средним на 1 час больше или на 1 час меньше, то p-value в обоих случаях = 0,46, значит гипотезу не отвергаем.
Почему одинаковые значения p-value как для +N, так и для -N, непонятно.

```python
# Считаем показатели t-статистики и p-value для двух выборок: часы работы с низкими и высокими зарплатами.
# Как интерпретировать результат, непонятно
from scipy.stats import ttest_ind
stat, p = ttest_ind(hr_high_salary_hours, hr_low_salary_hours)
stat, p
```

### 9 Разделить данные на тестовую и обучающую выборки. Построить модель LDA, предсказывающую уволился ли сотрудник на основе имеющихся факторов (кроме department и salary). Оценить качество модели на тестовой выборки 20

```python
# Готовим данные для LDA.
# Х - это датафрейм с количественными значениями (кроме столбца 'left')
# Для LDA независимые переменные должны быть непрерывными, поэтому не берем в выборку для LDA показатели:
# 'Work_accident', 'promotion_last_5years', которые принимают значения 0 или 1, т.е. имеют признаки категориальных переменных.
# Y - это ряд данных (серия), уволился или нет сотрудник - зависимая категориальная переменная
# Переводим y в массив numpy, иначе получаются NaN в итоговой таблице
X = hr_satisf.filter(items = ['satisfaction_level', 'last_evaluation', 'number_project', 'average_montly_hours', 
                           'time_spend_company'])
# X = hr_satisf.filter(items = ['satisfaction_level', 'last_evaluation', 'number_project', 'average_montly_hours', 
#                            'time_spend_company', 'Work_accident', 'promotion_last_5years'])
y = np.array(hr_satisf['left'])
```

```python
# Разделяем выборку на тренировочную и тестовую
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
y_test
```

```python
# Загружаем в функцию LDA тренировочную выборку.
# Строим прогноз Y_test по выборке X_test
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
lda_hr = LinearDiscriminantAnalysis()
lda_hr.fit(X_train, y_train)
lda_hr.predict(X_test)
```

```python
# Смотрим разницу факта и прогноза
# Добавлены названия колонок для интерпретации
d = {'y_test': y_test, 'y_predict': lda_hr.predict(X_test), 'y_test - y_predict': y_test-lda_hr.predict(X_test)}
result = pd.DataFrame(data=d)
result.iloc[0:20]
```

```python
# То же самое, в другой записи
result2 = pd.DataFrame([y_test, lda_hr.predict(X_test)]).T
result2.iloc[0:20]
```

```python
# Делаем расчет точности - отношение верных прогнозов к общему количеству позиций
# является хорошей метрикой для сбалансированных классов (наблюдений в категориях примерно одинаковое количество)
from sklearn.metrics import accuracy_score
accuracy_score(y_test, lda_hr.predict(X_test))
```

```python
# Коэффициенты дискриминатных линий - невысокие
lda_hr.coef_
```
